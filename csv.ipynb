```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
line = pd.read_csv("df_linear_totalTimePlayed.csv")
adap = pd.read_csv("df_adaptive_totalTimePlayed.csv")
```

```python
fig, (lineax, adapax) = plt.subplots(1, 2, figsize = (12,5), sharey=True, sharex=True)

line.rename(columns={"total time played" : "Linear storytelling"}).hist(grid = False, ax = lineax, color="#3DB1E5", bins=12, zorder=1, label="Answers")
adap.rename(columns={"total time played" : "Adaptive storytelling"}).hist(grid = False, ax = adapax, color="#E99B96", bins=12, zorder=1, label="Answers")

linemean = line["total time played"].mean()
adapmean = adap["total time played"].mean()


lineax.axvline(linemean, label="Average (%.2f)" %linemean, color="#444444", alpha=0.8, lw=1.5)
adapax.axvline(adapmean, label="Average (%.2f)" %adapmean, color="#444444", alpha=0.8, lw=1.5)

for i in range(0, 14, 2):
    lineax.axhline(i, dashes=[10, 5], alpha=0.8, color="#999999", lw=0.5, zorder=0)
    adapax.axhline(i, dashes=[10, 5], alpha=0.8, color="#999999", lw=0.5, zorder=0)

for i in range(0, 1200, 200):
    lineax.axvline(i, dashes=[10, 5], alpha=0.8, color="#999999", lw=0.5, zorder=0)
    adapax.axvline(i, dashes=[10, 5], alpha=0.8, color="#999999", lw=0.5, zorder=0)

lineax.set(xlabel="time spent (seconds)",
           ylabel="frequency of answers (n)",
           yticks=np.arange(0, 14, 2),
           xticks=np.arange(0, 1200, 200))

adapax.set(xlabel="time spent (seconds)",
           yticks=np.arange(0, 14, 2),
           xticks=np.arange(0, 1200, 200))

lineax.legend(loc=1)
adapax.legend(loc=1)

adapax.tick_params(axis='y', left=False, right=True, labelleft=False, labelright=True)

plt.suptitle("Total time spent", size=14, weight="semibold")
plt.tight_layout(pad=1.2, w_pad=0.5, h_pad=1.2)
plt.subplots_adjust(top=0.85)

fig.show()
```
